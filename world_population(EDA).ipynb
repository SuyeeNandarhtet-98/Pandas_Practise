Correlation

```python
df_numeric = df.select_dtypes(include=['number'])  # Select only numeric columns
df_numeric.corr()  # Compute correlation

```

```python
sns.heatmap(df_numeric.corr(), annot = True, cmap=sns.color_palette("pastel", as_cmap=True), linewidths=0.4)
plt.rcParams['figure.figsize'] =(20,7)
plt.show()
```

Data Grouping


```python
numeric_cols = df.select_dtypes(include='number').columns  # Select only numeric columns
numeric_cols
df.groupby('Continent')[numeric_cols].mean()
```

```python
df.columns
```

```python
df1 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by='2022 Population', ascending=False)
```

```python
df1
```

```python
df2= df1.transpose()
df2
```

```python
df2.plot()
```

```python

ax = df2.plot(marker='o', linestyle='-', figsize=(13, 6), cmap='Pastel1')
ax.set_title('Continent-wise Population Metrics', fontsize=16)
ax.set_xlabel('Continent', fontsize=12)
ax.set_ylabel('Values', fontsize=12)
plt.tight_layout()
plt.show()
```
